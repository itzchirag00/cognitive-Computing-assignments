```python
import numpy as np
import matplotlib.pyplot as plt

N=float(input("Enter the value of N: "))
a=np.linspace(-15,15,300)
b1=N*a**2
b2=N*np.cos(a)

plt.plot(a, b1, 'g-', label=f'b = {N}aÂ²')
plt.plot(a, b2, 'm--', label=f'b = {N}cos(a)')

plt.title(f'Mathematical Functions with N = {N}')
plt.xlabel('a')
plt.ylabel('b')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import random

info = {
'Topics': ['History', 'Geography', 'Biology', 'Art', 'Literature']}

df2 = pd.DataFrame(info)
df2['Grades'] = [random.randint(50, 90) for i in range(5)]

sns.barplot(x='Topics', y='Grades', data=df2, palette='coolwarm')

for i, grade in enumerate(df2['Grades']):
    plt.text(i, grade, str(grade), ha='center', va='bottom')

plt.title('Topic Grades Distribution')

plt.xlabel('Topics')

plt.ylabel('Grades')
plt.grid(True, axis='y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
student_id = 102317140
np.random.seed(student_id)
values = np.random.randn(50)

fig, ((plot1, plot2), (plot3, plot4)) = plt.subplots(2, 2, figsize=(12, 10))

cumulative_sum = np.cumsum(values)
plot1.plot(cumulative_sum, 'g-')
plot1.set_title('Cumulative Sum')
plot1.grid(True)
plot1.set_xlabel('Index')
plot1.set_ylabel('Cumulative Sum')

noise_values = np.random.normal(0, 0.01, 50)
plot2.scatter(range(50), values + noise_values, alpha=0.6)
plot2.set_title('Scatter Plot with Noise')
plot2.grid(True)
plot2.set_xlabel('Index')
plot2.set_ylabel('Value')

plot3.remove()
plot4.remove()

plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df_sales = pd.read_csv("https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv")


plt.figure(figsize=(8, 6))
plt.plot(df_sales['month_number'], df_sales['total_profit'], marker='o', linestyle='-', color='b')
plt.title("Total Profit Over Months")
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.grid()
plt.show()


plt.figure(figsize=(10, 6))
for column in df_sales.columns[1:-1]:
    plt.plot(df_sales['month_number'], df_sales[column], marker='o', label=column)
plt.title("Product Sales Over Months")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid()
plt.show()


plt.figure(figsize=(10, 6))
df_sales.set_index('month_number').plot(kind='bar', figsize=(10, 6))
plt.title("Company Sales Data")
plt.xlabel("Month")
plt.ylabel("Values")
plt.grid()
plt.show()
```
